```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
infilelist=glob.glob('EPIC*extended_details.txt')
```

```python
plt.figure(figsize=(15,7.5))
for index, each in enumerate(infilelist):
    #if index<100:
    if True:
        df=pd.read_csv(each,sep=r"\s*",skiprows=5)
        avals=df['AVALS']
        plt.plot(avals,c='b',alpha=0.2)
plt.ylim([0,2])
```
